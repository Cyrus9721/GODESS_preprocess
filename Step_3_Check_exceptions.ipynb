```python
def helper_incorrect_match(pdb_index, pdb_name, pdb_f, three_letter_val, accurate_mono_val):
    
    for j in range(len(pdb_f)):

        current_mono_three_ele = pdb_f.loc[j, :]['Residual_name']
        current_mono_ele = pdb_f.loc[j, :]['Residual_accurate_name']

        three_letter_list.append(current_mono_three_ele)
        actual_residue_list.append(current_mono_ele)

        if (current_mono_three_ele == three_letter_val) and (current_mono_ele == accurate_mono_val):
            print(three_letter_val, accurate_mono_val, pdb_index, pdb_f_name)
            return None
```

```python
def helper_check_correct_atoms(pdb_index, pdb_name, pdb_f, atom_name = 'C1'):
    res_num = np.unique(pdb_f['Residual_num'].values)
    for j in res_num:
        atom_list = pdb_f.loc[pdb_f['Residual_num'] == j, :]['Atom_name'].values
        appearance = list(atom_list).count(atom_name)
        if appearance >= 2:
            print('more than one ' + atom_name + ', residual No.:', j, 'pdb name:', pdb_name)
```
