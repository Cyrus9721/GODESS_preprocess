### Examine the label information

```python
mono_out_dictionary = 'dataset/Godess_final_data/mono_dictionary/dict_mono.json'
```

```python
labeled_files = os.listdir(label_dir)
monosaccharide_list
```

```python
mono_dict = {key: [] for key in monosaccharide_list}
duplicate_name_list = []
residue_duplicate_name_list_1 = []
residue_duplicate_linkage_list_1 = []
residue_duplicate_name_list_2 = []
residue_duplicate_linkage_list_2 = []


for i in tqdm(range(len(pdb_files))):
# for i in [1]:
    df_pdb_name = pdb_files[i]

    df_label_name = df_pdb_name.split('labeled_all_')[1].replace('.pdb', '')
    df_label_folder_name = int(df_label_name.replace('.csv', ''))
    df_label_path = os.path.join(label_dir, df_label_name)
    
    df_label = pd.read_csv(df_label_path)
    
    df_label_residue = df_label['Residue'].values
    
    # create dictionary for monosaccharde appearance 
    for res in df_label_residue:
        if res in monosaccharide_list:
            mono_dict[res].append(df_label_folder_name)
            mono_dict[res] = list(set(mono_dict[res]))
            
    # examine labeling issues
    df_label = df_label[~df_label['Linkage'].isna()]
    
    duplicate_rows = df_label[df_label.duplicated(subset=['Linkage'], keep=False)]
    
    if not duplicate_rows.empty:
        duplicate_name = df_pdb_name
        duplicate_label_df = duplicate_rows
        duplicate_label_df.index = range(len(duplicate_label_df))
        
        for j in range(len(duplicate_label_df)):
            current_linkage = duplicate_label_df['Linkage'].values[j]
            current_residue = duplicate_label_df['Residue'].values[j]

            duplicate_label_df_subset = duplicate_label_df.loc[duplicate_label_df['Linkage'] == current_linkage]

            duplicate_label_df_subset.index = range(len(duplicate_label_df_subset))

            even_rows = duplicate_label_df_subset.iloc[::2]

            residue_duplicate_name_list_1.extend(list(even_rows['Residue'].values))

            residue_duplicate_linkage_list_1.extend(list(even_rows['Linkage'].values))

            odd_rows = duplicate_label_df_subset.iloc[1::2]

            residue_duplicate_name_list_2.extend(list(odd_rows['Residue'].values))

            residue_duplicate_linkage_list_2.extend(list(odd_rows['Linkage'].values))

            duplicate_name_list.extend(list(np.repeat(duplicate_name, len(odd_rows['Linkage'].values))))
```

```python
# save duplicate mono df
df_find_messing_up = pd.DataFrame([residue_duplicate_name_list_1,
                                      residue_duplicate_linkage_list_1,
                                      residue_duplicate_name_list_2,
                                      residue_duplicate_linkage_list_2, 
                                      duplicate_name_list]).T
df_find_messing_up = df_find_messing_up.drop_duplicates(ignore_index = True)
df_find_messing_up.columns = ['residue_1', 'residue_1_linkage', 'residue_2', 'residue_2_linkage', 'glycan_name']
df_find_messing_up.to_csv('Messing_up_residues.csv', index = False)
```
